**Handling Null values**

````python
# Because the number of garage cars is categorial (1,2,3,4) we can replace the missing value by the most repeated 
most_repeated=x_test['GarageCars'].mode()
x_test.fillna(value=most_repeated,inplace=True)
# Because garage area is non categorial, we can fill the missing area by the average area
ave=x_test['GarageArea'].mean()
x_test.fillna(value=ave,inplace=True)
````

````python
x_test.isnull().sum()
````

Removing outliers
---

````python
fig,ax =plt.subplots(2,1,figsize=(12,9))
sns.boxplot(data=x_train,ax=ax[0])
sns.boxplot(data=y_train,ax=ax[1])
````

```
From the data, points outside the upper and lower wiskers from the boxplot are considered to be outliers and must be removed 
```

````python
import copy
def drop_outliers(df):
    df= copy.deepcopy(df)
    Q1=df.quantile(0.25)
    Q3=df.quantile(0.75)
    IQR=Q3-Q1
    lwr_bound=Q1-1.5*IQR
    hgh_bound=Q3+1.5*IQR
    indexes=[]
    for col in df.columns:
        temp=df[(df[col] < lwr_bound[col])| (df[col]>hgh_bound[col])].index
        df=df.drop(temp)
    return df


temp_df= pd.concat([x_train, y_train],axis=1)
temp_df=drop_outliers(temp_df)

fig,ax =plt.subplots(2,1,figsize=(12,9))
sns.boxplot(data=temp_df[features],ax=ax[0])
sns.boxplot(data=temp_df[target],ax=ax[1])
````

````python
sns.pairplot(data=temp_df,x_vars=features,y_vars=target)
````

```
    Looking at the boxplots and pair plots it can be seen that the new training data has no outlier points
```

````python
import xgboost as xgb
model =xgb.XGBRegressor()
````

````python
x_train=temp_df[features]
y_train=temp_df[target]
print(x_train.shape)
model= xgb.XGBRegressor()
model.fit(x_train,y_train)

predictions=model.predict(x_test)
# model.to_csv('predictions.csv',index=False)
predictions=pd.Series(predictions,name='SalePrice')
df=pd.concat([df_test['Id'],predictions],axis=1,)
df.to_csv('predictions.csv',index=False)
````
